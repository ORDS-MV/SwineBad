```python
import re
import os
```

```python
years = [int(name) for name in os.listdir("./fulltexts") if os.path.isdir(os.path.join('./fulltexts',name))]
years.sort()

years_files = {}
for year in years:
    fpath = os.path.join("./fulltexts",str(year))
    years_files[year] = sorted([name for name in os.listdir(fpath) if os.path.isfile(os.path.join(fpath,name))])
    
known_source_loc = ['Berlin', 'Stettin']
```

```python
#fn = "./fulltexts/1915/swinba_636776093_1915_06_05_00000005.xml"
year = 1915

all_year_data = {}
    
for year in years_files.keys():
    year_files = years_files[year]
    days = {}
    for f in year_files:
        f = os.path.join("./fulltexts",str(year),f)
        in_data = False
        with open(f,'r') as file:
            loc_for_day = {l:0 for l in known_source_loc}
            for line in file:
                if line.startswith("Name, Stand, Wohnort und hiesige Wohnung"):
                    # data starts next line
                    in_data = True
                    day += 1
                    continue
                if line.startswith("Summe"):
                    # data stopped last line
                    in_data = False
                    break
                for loc in known_source_loc:
                    if loc in line:
                        loc_for_day[loc] += 1
            days[f] = loc_for_day
    all_year_data[year] = days
```
